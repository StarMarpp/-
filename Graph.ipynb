# BFS графа

Дан ориентированный граф. Задача состоит в том, чтобы выполнить обход этого графа в ширину, начиная с 0.
Примечание. Переход от узла u к узлу v возможен только в том случае, если есть ребро от u до v. Найдите обход BFS графа, начиная с 0-й вершины, от слева направо в соответствии с входным графиком. Кроме того, вам следует учитывать только узлы, прямо или косвенно связанные с узлом 0.

**Входные данные**:
V = 5, E = 4
adj = {{1,2,3},{},{4},{},{}}

**Пояснение**:



**Выходные данные**: 0 1 2 3 4
0 подключен к 1, 2, 3.
2 подключен к 4.
поэтому, начиная с 0, он перейдет к 1, затем к 2
затем 3. После этого от 2 до 4, таким образом, bfs будет
0 1 2 3 4.

```python
from typing import List
from queue import Queue
from collections import deque
class Solution:
    #Function to return Breadth First Traversal of given graph.
    def bfsOfGraph(self, V: int, adj: List[List[int]]) -> List[int]:
        ans = []
        q = deque([0])
        visited = [0] * V
        
        while q:
            v = q.popleft()
            if visited[v] == 0:
                ans.append(v)
            visited[v] = 1
            
            for u in adj[v]:
                if visited[u] == 0:
                    q.append(u)
    
        return ans
```
