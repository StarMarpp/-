# Подсчет листьев в двоичном дереве

Учитывая двоичное дерево размера N , вам нужно посчитать в нем листья. Например, в следующем дереве есть два листа



```python
def countLeaves(root):
    if not root.right and not root.left:
        return 1
    ans = 0
    if root.left:
        ans += countLeaves(root.left)
    if root.right:
        ans += countLeaves(root.right)
    return ans #найти сумму всех деревьев
```

# Правый вид двоичного дерева

Учитывая двоичное дерево, найдите  правильное представление о нем. Правый вид двоичного дерева — это набор узлов, видимых при просмотре дерева с правой стороны.

Вид справа на следующее дерево: 1 3 7 8.



```python
Right view of following tree is 1 3 7 8.

#           1
#       /     \
#      2        3
#   /   \      /    \
#   4     5   6    7
#     \
#      8


'''
# Node Class:
class Node:
    def init(self,val):
        self.data = val
        self.left = None
        self.right = None
'''

def height(root):
    if root:
        return max(height(root.left), height(root.right)) + 1
    return 0

def curLevel(root, lev):
    if root:
        if lev == 1:
            return [root.data]
        return curLevel(root.left, lev-1) + curLevel(root.right, lev-1)
    return []

class Solution:
    #Function to return list containing elements of right view of binary tree.
    def rightView(self,root):
        h = height(root)
        ans = []
        for i in range(1, h+1):
            ans.append(curLevel(root, i)[-1])
        return ans
```

# Самый низкий общий предок в BST

Дано двоичное дерево поиска (все значения уникальны) и два значения узлов n1 и n2 (n1!=n2). Найдите самых низких общих предков двух узлов в BST.

Вход:



n1 = 7, n2 = 8 

**Выход**: 7

```python
def LCA(root, n1, n2):
    #пусть значение равно корню
    if n1 < root.data and n2 < root.data:
        return LCA(root.left, n1, n2)
    elif n1 > root.data and n2 > root.data:
        return LCA(root.right, n1, n2)
    else:
        return root
```

# Обход порядка уровней в форме спирали

Дано двоичное дерево, и задача состоит в том, чтобы найти порядок обхода дерева по спирали.

Среднее значение обхода спирального порядка: начиная с уровня 0 для корневого узла, для всех четных уровней мы печатаем значение узла справа налево, а для всех нечетных уровней мы печатаем значение узла слева направо. 

Для дерева ниже функция должна возвращать 1, 2, 3, 4, 5, 6, 7.



```python
def findSpiral(root):
    lvl =[root]
    ans = []
    dir = -1
    
    while lvl:
        for v in lvl[::dir]:
            ans.append(v.data)
                
        n_lvl = []
        
        for v in lvl:
            if v.left:
                n_lvl.append(v.left)
            if v.right:
                n_lvl.append(v.right)
        lvl = n_lvl
        dir = - dir
    return ans
```

# Бинарный поиск в лесу

**Вход**:

n = 5

k = 4

дерево[] = {2, 3, 6, 2, 4}

**Результат**: 3

**Пояснение**: Древесина, собранная путем рубки деревьев на высоте 3 = 0 + 0 + (6-3) + 0 + (4-3) = 4 следовательно, 3 следует вычесть из всех чисел. Поскольку 2 меньше 3, ничего не получается, вычитается из него.

```python
class Solution:
    def find_height(self,tree,n,k):        
        low = 0
        high = max(tree)
        h = (low + high) // 2
        while low <= high:
            wood = sum([t - h for t in tree if t > h])
            if wood == k:
                return h
            elif wood < k:
                high = h - 1
            else:
                low = h + 1
            h = (low + high) // 2
        
        return -1
```
